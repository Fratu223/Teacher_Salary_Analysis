```python
# Importing the needed packages

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Plotting the 50 states + District of Columbias' average teacher pay on a column bar chart

df_teacher_pay = pd.read_csv('../Data/teacher-pay-by-state-2024.csv')

plt.figure(figsize=(10, 6))
plt.bar(df_teacher_pay['state'], df_teacher_pay['TeacherPay_AnnualMeanSalary_USD_2024'], color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Annual Mean Salary (USD)')
plt.title('Teacher Pay - Annual Mean Salary (2024)')
plt.tight_layout()
plt.show()
```

```python
# Plotting the 50 states + District of Columbias' cost of living index on a column bar chart

df_cost_of_living = pd.read_csv('../Data/cost-of-living-index-by-state-2024.csv')

plt.figure(figsize=(10, 6))
plt.bar(df_cost_of_living['state'], df_cost_of_living['CostOfLivingIndex2023'], color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Cost of Living Index')
plt.title('Cost of Living Index (2023)')
plt.tight_layout()
plt.show()
```
