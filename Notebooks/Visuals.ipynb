```python
# Importing the needed packages

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Plotting the 50 states + District of Columbias' average teacher pay on a column bar chart

df = pd.read_csv('../Data/teacher-pay-by-state-2024.csv')

plt.figure(figsize=(10, 6))
plt.bar(df['state'], df['TeacherPay_AnnualMeanSalary_USD_2024'], color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Annual Mean Salary (USD)')
plt.title('Teacher Pay - Annual Mean Salary (2024)')
plt.tight_layout()
plt.show()
```
