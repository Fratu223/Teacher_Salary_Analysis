```python
# Importing the needed packages

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Plotting the 50 states + District of Columbias' average teacher pay on a column bar chart

df_teacher_pay = pd.read_csv('../Data/teacher-pay-by-state-2024.csv')

plt.figure(figsize=(10, 6))
plt.bar(df_teacher_pay['state'], df_teacher_pay['TeacherPay_AnnualMeanSalary_USD_2024'], color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Annual Mean Salary (USD)')
plt.title('Teacher Pay - Annual Mean Salary (2024)')
plt.tight_layout()
plt.show()
```

```python
# Plotting the 50 states + District of Columbias' cost of living index on a column bar chart

df_cost_of_living = pd.read_csv('../Data/cost-of-living-index-by-state-2024.csv')

plt.figure(figsize=(10, 6))
plt.bar(df_cost_of_living['state'], df_cost_of_living['CostOfLivingIndex2023'], color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Cost of Living Index')
plt.title('Cost of Living Index (2023)')
plt.tight_layout()
plt.show()
```

```python
# Plotting the normalized teacher annual mean salary in USD vs the cost of living index by state for each every state + District of Columbia

df_combined = pd.merge(df_teacher_pay, df_cost_of_living, on='state')

df_combined['NormalizedTeacherPay'] = (df_combined['TeacherPay_AnnualMeanSalary_USD_2024'] - df_combined['TeacherPay_AnnualMeanSalary_USD_2024'].min()) / (df_combined['TeacherPay_AnnualMeanSalary_USD_2024'].max() - df_combined['TeacherPay_AnnualMeanSalary_USD_2024'].min())
df_combined['NormalizedCostOfLiving'] = (df_combined['CostOfLivingIndex2023'] - df_combined['CostOfLivingIndex2023'].min()) / (df_combined['CostOfLivingIndex2023'].max() - df_combined['CostOfLivingIndex2023'].min())

plt.figure(figsize=(14, 8))
plt.bar(df_combined['state'], df_combined['NormalizedTeacherPay'], color='blue', alpha=0.7, label='Normalized Annual Mean Salary (USD)')
plt.plot(df_combined['state'], df_combined['NormalizedCostOfLiving'], color='red', marker='o', label='Normalized Cost of Living')

plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Normalized Values (0 - 1)')
plt.title('Normalized Annual Mean Salary (USD) vs Cost of Living Index by State')

plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Plotting the normalized teacher annual mean salary in USD vs the cost of grocery index by state for each every state + District of Columbia

df_combined['NormalizedCostOfLivingGrocery'] = (df_combined['CostOfLivingIndexGroceryCostsIndex'] - df_combined['CostOfLivingIndexGroceryCostsIndex'].min()) / (df_combined['CostOfLivingIndexGroceryCostsIndex'].max() - df_combined['CostOfLivingIndexGroceryCostsIndex'].min())

plt.figure(figsize=(14, 8))
plt.bar(df_combined['state'], df_combined['NormalizedTeacherPay'], color='blue', alpha=0.7, label='Normalized Annual Mean Salary (USD)')
plt.plot(df_combined['state'], df_combined['NormalizedCostOfLivingGrocery'], color='green', marker='o', label='Normalized Cost of Grocery')

plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Normalized Values (0 - 1)')
plt.title('Normalized Annual Mean Salary (USD) vs Cost of Grocery Index by State')

plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
